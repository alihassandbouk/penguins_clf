```python
class encodeing(BaseEstimator ,TransformerMixin ):
    def fit(self,X , y= None):
        return self
    
        
    def transform(self,X):
        encoder = OneHotEncoder(handle_unknown="ignore")
        matrix = encoder.fit_transform(X[['sex']]).toarray()
        col = encoder.get_feature_names_out()
        for i in range(len(matrix.T)):
            X[col[i]] = matrix.T[i]
            
        matrix = encoder.fit_transform(X[['island']]).toarray()
        col = encoder.get_feature_names_out()
        for i in range(len(matrix.T)):
            X[col[i]] = matrix.T[i]
            
        return X.drop(["sex" , "island"] , axis = 1)
            
```

```python
encoder = encodeing()
```
